```python
# Requisições por número da página

df_cnpj = pd.DataFrame()
pagina = 0

iniciar_em = 13364

for numero in range (iniciar_em, iniciar_em+10001):
    html_cnpj = pd.read_html("https://brasil.io/dataset/documentos-brasil/documents?page="+str(numero))[0]
    html_cnpj = html_cnpj[html_cnpj["Tipo de Documento"]=="CNPJ"]
    df_cnpj = pd.concat([df_cnpj, html_cnpj]
                        , axis=0
                        , join="outer")
    pagina = numero
```

```python
print(f"""Tamanho do df_cnpj: {df_cnpj.shape}
Parei na página: {pagina}""")
```

```python
df_cnpj.to_csv("df_cnpj_"+str(numero)+".csv")
```
