## 1) Tratando o df para ser consumido pelo modelo de árvore de decisão

Lembrete importante, o DecisionTreeClassifier não aceita valores que não numéricos. Portanto, é importante segui para uma estratégia de transformação das variáveis categóricas em Dummy Viariables:

```python
# Copiando df original para um novo df (df_tratado) e substituindo os valores negativos pela média dos positivos.

df_tratado = df.copy()

df_tratado.replace(to_replace="?", value=np.nan, inplace=True)

lista_atributos_categoricos = list(df_tratado.select_dtypes(exclude=["int64", "float64"]).columns)

state_dummies = pd.get_dummies(df_tratado[lista_atributos_categoricos], drop_first=True)
df_tratado.drop(lista_atributos_categoricos, axis=1, inplace=True)

df_tratado = pd.concat([df_tratado, state_dummies], axis=1)

df_tratado.sample(10).T
```

## 2) Aplicação da Aprendizagem por árvores de decisão

### 2.1) Usando *train_test_split*  

```python
# Separando os fatores/previsores da resposta/target:
    
x = df_tratado.iloc[:, 0:-1]
y = df_tratado.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier, export

classificador = DecisionTreeClassifier(criterion='entropy', random_state=0, max_depth=len(x.columns)**0.5)
classificador.fit(x_train, y_train)
```

```python
print(classificador.feature_importances_)
```

```python
classificador.max_depth
```

```python
y_pred = classificador.predict(x_test)
```

### Verificando o como o modelo se saiu

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

print("Métrica 1 (Acurácia): {:.6}.".format(
    accuracy_score(y_test, y_pred, normalize=True, sample_weight=None)))
print("Métrica 2 (Relatório de classificação):\n{}.".format(
    classification_report(y_test, y_pred)))
print("Métrica 2 (Relatório de classificação):\n{}.".format(
    confusion_matrix(y_test, y_pred)))
```

```python
plot_confusion_matrix_2(y_test, y_pred, normalize=True);
```

### Agora, usando validação cruzada.

```python
classificador = DecisionTreeClassifier(criterion='entropy', max_depth=2)
classificador.fit(x, y)

resultados = cross_val_score(classificador, x, y, cv = 10)

classe = y.values

print("Acurácia média: {:.3} ± {:.3}.".format(resultados.mean(), resultados.std()))
```

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import tree
from sklearn.datasets import load_wine
from IPython.display import SVG
from graphviz import Source
from IPython.display import display

dot_data = Source(tree.export_graphviz(classificador, out_file=None,
                                       feature_names = list(x.columns),
                                       class_names = ["0", "1"],
                                       filled = True,
                                       leaves_parallel=True))

display(SVG(dot_data.pipe(format='svg')))
```

```python
from ipywidgets import interactive

def plot_tree(crit, split, depth, min_split, min_leaf=0.2):
    classificador = DecisionTreeClassifier(criterion=crit
                                           , splitter=split
                                           , random_state=0
                                           , max_depth=depth
                                           , min_samples_split=min_split
                                           , min_samples_leaf=min_leaf)
    classificador.fit(x_train, y_train)
    
    graph = Source(tree.export_graphviz(classificador
      , out_file=None
      , feature_names=list(x.columns)
      , class_names = ["0", "1"]
      , filled = True))
   
    display(SVG(graph.pipe(format='svg')))
    
    return classificador 

inter=interactive(plot_tree
                  , crit = ["gini", "entropy"]
                  , split = ["best", "random"]
                  , depth=[1,2,3,5,8,13,21]
                  , min_split=(0.1,1)
                  , min_leaf=(0.1,0.5))

display(inter)
```
